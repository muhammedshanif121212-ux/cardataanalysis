## Data Cleaning

- Remove text units and convert to numbers

> Changing datatypes

```python
df['mileage'] = pd.to_numeric(
    df['mileage']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None)
)
df
```

```python
df['max_power'] = pd.to_numeric(
    df['max_power']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None)
)
df
```
